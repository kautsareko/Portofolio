```python
import pandas as pd

dcrawl= pd.read_csv("B:\Tugas akhir\data csv\DataYangDiolah.csv",encoding = 'utf8', low_memory=False)
dcrawl['product_name']=dcrawl['product_name'].str.lower()
dcrawl['product_brand']=dcrawl['product_brand'].str.lower()
dcrawl['product_name']=dcrawl['product_name'].replace(to_replace=r'\r\r\n', value='', regex=True)
dcrawl['product_brand']=dcrawl['product_brand'].replace(to_replace=r'\r\r\n', value='', regex=True)
dcrawl.replace('^\s+', '', regex=True, inplace=True)
dcrawl.replace('\s+$', '', regex=True, inplace=True)
dcrawl.sort_values(by=['product_name'], inplace=True)
dcrawl.to_csv("dcrawlSorting.csv")

halal= pd.read_csv("B:\Tugas akhir\data csv\Halalfoodproductsyangdihapusblank.csv",encoding = 'utf8', low_memory=False)
halal['ProductHalal']=halal['ProductHalal'].str.lower()
halal['ManufakturHalal']=halal['ManufakturHalal'].str.lower()
halal['ProductHalal']=halal['ProductHalal'].replace(to_replace=r'\r\r\n', value='', regex=True)
halal['ManufakturHalal']=halal['ManufakturHalal'].replace(to_replace=r'\r\r\n', value='', regex=True)
halal.replace('^\s+', '', regex=True, inplace=True)
halal.replace('\s+$', '', regex=True, inplace=True)
halal.sort_values(by=['ProductHalal'], inplace= True)
halal.to_csv("halalSorting.csv")
dcrawl
```
