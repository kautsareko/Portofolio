# Abaikan kodingan ini

```python
import pandas as pd
df=pd.read_csv("datasama.csv")
df['Products Listing']=df['Products Listing'].replace(to_replace=r'\r\r\n', value='', regex=True)
df['Manufaktur']=df['Manufaktur'].replace(to_replace=r'\r\r\n', value='', regex=True)
df.replace('^\s+', '', regex=True, inplace=True)
df.replace('\s+$', '', regex=True, inplace=True)
# df=pd.Series()
# df=df.str.title()
df['Products Listing']=df['Products Listing'].str.title()
df['Manufaktur']=df['Manufaktur'].str.title()
df.to_csv("datasamafix.csv")
```
