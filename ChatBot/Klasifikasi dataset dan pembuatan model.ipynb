```python
import csv
import pickle


class Kata:
    
    namaKata=""
    jumlahKata=""
    pKata=""
    
    def __init__(self, nama):
        self.namaKata=nama
        self.jumlahKata=1
        self.pKata=0
    
class Category:
    
    namaCategory=""
    listKata=[]
    pCategory=0
    jumlahCategory=0
    totalJumlahCategory=0
    jumlahTotalKata=0
    
    def __init__(self,nama,jumlah,kata):
        self.namaCategory=nama
        self.listKata=[Kata(kata)]
        self.pCategory=0
        self.jumlahCategory=1
        self.totalJumlahCategory=jumlah
        self.jumlahTotalKata=1
        
    def addListWord(self, word):
        self.jumlahTotalKata+=1
        
        cek=True
        for add1 in self.listKata:
            
            if add1.namaKata==word:
                add1.jumlahKata+=1
                cek=False
                break
        if cek:
            self.listKata.append(Kata(word))
    
    def calculateP(self):
        
        self.pCategory=self.jumlahCategory/self.totalJumlahCategory
        
        
        for cal1 in self.listKata:
           
            cal1.pKata=float((cal1.jumlahKata+1)/(self.jumlahTotalKata+len(self.listKata)))
         
    
    def getKataP(self, kata1):
        for g1 in self.listKata:
            if kata1==g1.namaKata:
                return g1.pKata
        
        return 1/(self.jumlahTotalKata+len(self.listKata))
    
    def cetak(self):
        tesy=0
        print(len(self.listKata))
        print(self.namaCategory)
        for ss in self.listKata:
            print(ss.namaKata)
            print(ss.jumlahKata)
            print(ss.pKata)
            tesy+=ss.pKata
        print(tesy)
    
       
jumlahbaris=0
dataTraining=[]
dataTesting=[]
dataPercakapan=[]
listCat=[]
jumlahPertanyaan=0

with open('data.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        dataPercakapan.append(row);
        if row[0]=="pembeli":
            jumlahPertanyaan+=1
   
        jumlahbaris+=1
        
    

index=0
jTraining=int(jumlahPertanyaan*8/10)

for sss in dataPercakapan:
    
    if sss[0]=="pembeli":
    
        if index<jTraining:
            dataTraining.append(sss)
        else:
            dataTesting.append(sss)
        index+=1
    
    
for x in dataTraining:
    
    splitPercakapan=x[1].split(" ")
    
    for y in splitPercakapan:
            
        if listCat==[]:
            listCat.append(Category(x[2],jTraining,y))
        else:
            cekCat=True
            for z in listCat:
                
                if z.namaCategory==x[2]:
                    z.addListWord(y)
                    cekCat=False
                    break
            if cekCat:
                listCat.append(Category(x[2],jTraining,y))
    z.jumlahCategory+=1

for k in listCat:
    k.calculateP()
    

#batas data training.............................................

akurasi=0


precision=[]
recall=[]
benar=[]

for ds in listCat:
    precision.append([ds.namaCategory,0])
    recall.append([ds.namaCategory,0])
    benar.append([ds.namaCategory,0])


for b in dataTesting:
    prediksi=0
    CategoryPrediksi=""
    
    
    for a in listCat:
        totalCat=a.pCategory
   
        
        pertanyaanTest=b[1].split(" ")
            
        for c in pertanyaanTest:
            totalCat+=a.getKataP(c)
    
        if totalCat>prediksi:
            
            prediksi=totalCat
            CategoryPrediksi=a.namaCategory

    
    for pre in precision:
        if pre[0]==CategoryPrediksi:
            pre[1]+=1
            break
    for rec in recall:
        if b[2]==rec[0]:
            rec[1]+=1
            break
            
    if CategoryPrediksi==b[2]:
        akurasi+=1
        for ben in benar:
            if b[2]==ben[0]:
                ben[1]+=1
                
                    

print("")    

akurasi=akurasi/len(dataTesting)
print("akurasi : ",akurasi)

print("")


for ben in benar:
    
    for pre in precision:
        if ben[0]==pre[0]:
            if pre[1]==0:
                 print("Precision ",pre[0]," : 0")
            else:
                print("Precision ",pre[0]," : ",(ben[1]/pre[1]))
print("")
for ben in benar:
    for rec in recall:
        if ben[0]==rec[0]:
            if rec[1]==0:
                 print("recall ",rec[0]," : 0")
            else:
                print("recall ",rec[0]," : ",(ben[1]/rec[1]))
            
    
    

    

#for s in listCat:
   # print(s.jumlahCategory, s.namaCategory)
                
###########batas testing..........................
pickle.dump( listCat, open( "save.p", "wb" ) )
pickle.dump( dataPercakapan, open( "save.w", "wb" ) )
```
