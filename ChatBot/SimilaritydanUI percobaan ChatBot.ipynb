```python
import csv
from tkinter import *
from tkinter.ttk import *
import pickle


class Kata:
    
    namaKata=""
    jumlahKata=""
    pKata=""
    
    def __init__(self, nama):
        self.namaKata=nama
        self.jumlahKata=1
        self.pKata=0
    
class Category:
    
    namaCategory=""
    listKata=[]
    pCategory=0
    jumlahCategory=0
    totalJumlahCategory=0
    jumlahTotalKata=0
    
    def __init__(self,nama,jumlah,kata):
        self.namaCategory=nama
        self.listKata=[Kata(kata)]
        self.pCategory=0
        self.jumlahCategory=1
        self.totalJumlahCategory=jumlah
        self.jumlahTotalKata=1
        
    def addListWord(self, word):
        self.jumlahTotalKata+=1
        
        cek=True
        for add1 in self.listKata:
            
            if add1.namaKata==word:
                add1.jumlahKata+=1
                cek=False
                break
        if cek:
            self.listKata.append(Kata(word))
    
    def calculateP(self):
        
        self.pCategory=self.jumlahCategory/self.totalJumlahCategory
        
        
        for cal1 in self.listKata:
           
            cal1.pKata=float((cal1.jumlahKata+1)/(self.jumlahTotalKata+len(self.listKata)))
         
    
    def getKataP(self, kata1):
        for g1 in self.listKata:
            if kata1==g1.namaKata:
                return g1.pKata
        
        return 1/(self.jumlahTotalKata+len(self.listKata))
    
    def cetak(self):
        tesy=0
        print(len(self.listKata))
        print(self.namaCategory)
        for ss in self.listKata:
            print(ss.namaKata)
            print(ss.jumlahKata)
            print(ss.pKata)
            tesy+=ss.pKata
        print(tesy)
    
       

dataPercakapan=pickle.load( open( "save.w", "rb" ) )
listCat=pickle.load( open( "save.p", "rb" ) )






class UI(Frame):
    
    def __init__(self):
        super().__init__()   
         
        self.initUI()

 
            
    def initUI(self):
        self.master.title("Dagang HP")
        self.style = Style()
        self.style.theme_use("default")
        
        
        
       
        
        
      
        self.pack(fill=BOTH, expand=True)
        frame2 = Frame(self)
        frame2.pack(fill=X)
        
        txt = Text(frame2)
        txt.pack(fill=BOTH, pady=5, padx=5, expand=True) 
        indexChat=float(txt.index("end"))+1
        txt.insert(indexChat,"Penjual : halo ada yang bisa kami bantu?\n")
        txt.config(state=DISABLED)
        indexChat=float(indexChat+1)
        
        frame1 = Frame(self)
        frame1.pack(fill=X)
        
        entry1 = Entry(frame1)
        entry1.pack(fill=X,side=LEFT, padx=5, pady=5, expand=True )
      
    
        
        
        def enter():

            Pertanyaan = entry1.get()
            p2="Pembeli :  "+Pertanyaan+"\n"
            txt.config(state=NORMAL)
            indexChat=float(txt.index("end"))+1
            txt.insert(indexChat,p2)
            
            entry1.delete(0,END)
        
           
        
            prediksi=0
            CategoryPrediksi=""

            for aa in listCat:
                totalCat=aa.pCategory
             

                pertanyaanTest=Pertanyaan.split(" ")

                for cc in pertanyaanTest:
                    totalCat+=aa.getKataP(cc)
                   
                if totalCat>prediksi:

                    prediksi=totalCat
                    CategoryPrediksi=aa.namaCategory

           
            Jawaban=""
            kesamaan=0
            tes1=False
            for ff in dataPercakapan:

                if tes1:
                    Jawaban=ff[1]
                    tes1=False

                sem=0

                if ff[0]=="pembeli":

                    if ff[2]==CategoryPrediksi:

                        a = set(Pertanyaan.split()) 
                        b = set(ff[1].split())
                        c = a.intersection(b)
                        sem=float(len(c)) / (len(a) + len(b) - len(c))

                        if sem>kesamaan:
                            kesamaan=sem
                            tes1=True

       
            indexChat=float(txt.index("end"))+1
           
            if Jawaban=="":
                txt.insert(indexChat,"Penjual : Mohon maaf kami tidak mengerti pertanyaan anda, bisa diulangi?\n")
            else:
                p3="penjual : "+Jawaban+"\n"
                txt.insert(indexChat,p3)
           
            txt.config(state=DISABLED)


              
      
       
       
        def enter1(event):
            enter()
        
        
        entry1.bind('<Return>', enter1)
        closeButton = Button(frame1, text="Enter", command=enter)
        closeButton.pack(side=RIGHT, padx=5, pady=5)
        
        
  


def main():
  
    root = Tk()
    root.geometry("500x500+300+300")
    app = UI()
    root.mainloop()  
    
    
   
   



if __name__ == '__main__':
    main()  
    

        
```
